```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bbc.com/news'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
headlines = soup.find_all('h1', class_='headline')
for idx, headline in enumerate(headlines, start=1):
    print(f"{idx}. {headline.get_text(strip=True)}")
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Send a GET request to the website
url = 'https://www.bbc.com/news' # Replace with the URL you're scraping
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Step 2: Extract the headlines
headlines = soup.find_all('h3')

# Step 3: Save the headlines to a text file
with open('headlines.txt', 'w', encoding='utf-8') as file:
    for idx, headline in enumerate(headlines, start=1):
        file.write(f"{idx}. {headline.get_text(strip=True)}\n")

print("Headlines saved to headlines.txt")
```
